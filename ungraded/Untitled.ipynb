```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
import matplotlib
%matplotlib inline

df1 = pd.read_csv('../data/raw/world_happiness_report.csv')

print(df1)
```

```python
print(df1['Year'].unique())
```

```python
df1 = df1[df1.Year != 2022]

df1
```

```python
df1 = df1.copy().drop(['Happiness Rank','Freedom','Trust (Government Corruption)', 'Generosity'], axis=1)
df1
```

```python
df1= df1.copy().drop(df1.columns[0], axis=1)
print(df1.shape)
df1
```

```python
df1["Happiness Score"]=df1["Happiness Score"].astype(float)
df1["Economy (GDP per Capita)"]=df1["Economy (GDP per Capita)"].astype(float)
df1["Family (Social Support)"]=df1["Family (Social Support)"].astype(float)
df1["Health (Life Expectancy)"]=df1["Health (Life Expectancy)"].astype(float)

```

```python
df1
```

```python
def normalize_with_percent(df):
    result = df1.copy()
    for feature_name in df1.columns:
        if df1[feature_name].dtype==float:
            # if dtype is float then normalize it
            max_value = df1[feature_name].max()
            min_value = df1[feature_name].min()
            result[feature_name] = ((df1[feature_name] - min_value) / (max_value - min_value))*100
    return result
```

```python
normalized_df1 = normalize_with_percent(df1)
print(normalized_df1)
```

```python
normalized_df1.to_csv('../data/processed/world_happiness_report_processed_Untitled.csv')
```
